## Doom: A Glimpse into Interdimensional Chaos The Doom franchise, a cornerstone of gaming, boasts a rich lore spanning ancient civilizations and modern-day conflicts. This story combines the original 1993 narrative with the 2004 reboot, Doom 3, offering a concise look at its universe. ### Ancient Mars and the First Cataclysm Long before humanity, a sophisticated Martian civilization mastered interdimensional travel. Their experiments, however, inadvertently opened a portal to Hell, unleashing demonic forces that annihilated them. The survivors fled, some to Earth, while the rest created the Soul Cube, a mystical artifact used to banish the demons and seal the Hell Gate. They left behind a warning about their dangerous technology before their extinction. ### Humanity's Unwitting Repeat Centuries later, humanity evolved and, in the early 2000s, established the Union Aerospace Corporation (UAC) to terraform Mars. Excavating the planet, they discovered the ancient Martian ruins and their portal technology. Unable to decipher the warnings, the UAC reactivated the dimensional portals, once again opening a gateway to Hell. Demons swarmed, decimating the UAC personnel. ### The Marine's Stand A lone Marine was dispatched to Mars to investigate the distress calls. He uncovered the UAC's illicit activities and the demonic invasion. Wielding the recovered Soul Cube, he battled through Mars bases, traversed dimensional portals, and ultimately confronted the demons at Hell's Gate. Using the Soul Cube's power, he sealed the portal, returning to Earth as a hero. ### Doom 2: A Sequel's Threat While Doom 3 served as a reboot, Doom 2 acted as a direct sequel to the original. Upon returning to Earth, the Marine discovered a new Hell Gate had opened. After a demonic incursion at a human evacuation facility, he learned the demons had breached a portal in his hometown. The Marine eliminated the threat there and entered Hell himself, defeating the Icon of Sin and permanently sealing the gate. Earth began its recovery, and the threat of demonic invasion was seemingly over.
